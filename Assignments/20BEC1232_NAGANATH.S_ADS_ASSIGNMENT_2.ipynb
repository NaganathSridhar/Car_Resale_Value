# HANDLING THE MISSING VALUES

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
data['age'].fillna(data['age'].mean(),inplace=True)
```

```python
mode = data['deck'].mode().values[0]
data['deck'].fillna(value=mode, inplace=True)
```

```python
mode = data['embarked'].mode().values[0]
data['embarked'].fillna(value=mode,inplace=True)
```

```python
mode = data['embark_town'].mode().values[0]
data['embark_town'].fillna(value=mode,inplace=True)
```
